```python
%%ampl_eval
reset;

var x1 >= 0;
var x2 >= 0;

maximize Z: 4 * x1 + 3 * x2;
s.t. R1: x1 + 3 * x2 <= 7;
s.t. R2: 2 * x1 + 2 * x2 <= 8;
s.t. R3: x1 + x2 <= 3;
s.t. R4:x2 <= 2;
```

```python
%%ampl_eval
option solver cbc;
solve;
display Z,x1,x2;
```
